```
# Outer Product 
import numpy as np 
a = np.array([2, 4, 8, 7, 7])
b = np.array([2, 3, 1, 7, 8]) #outer product in both order 
opab = np.outer(a,b)
opba = np.outer(b,a) 
print('\n\n---A---\n', a) 
print('\n\n---B---\n', b) 
print('\n\nOuter Product as A.B : ', opab) 
print('\n\nOuter Product as A.B : ', opba)
```

```
# Identity Matrix Property (I^k = I) 
import numpy as np 
I = np.eye(4) 
print("\n---I(4x4)---\n", I) 
k = 14 
Ik = I 
for i in range(14):Ik =I*Ik
print('\n\n--- I^k ----\n', Ik)
```

```
# Identity Matrix Property (AI = A) 
import numpy as np 
I = np.eye(3) 
print("---Matrix I---\n", I) 
A = np.array([[2,3,4], [3,45,8], [4,8,78]]) 
print("---Matrix A---\n", A) 
ai = np.matmul(A, I) 
print('Identity Matrix Property I.A', ai) 
if ai.all() == A.all(): 
  print("AI = A")
```

```
# Inverse of a Matrix 
import numpy as np 
M = np.array([[2, 3, 4], [3, 5, 8], [4, 8, 7]]) 
print("---Matrix A---\n", M) 
ai = np.linalg.inv(M) 
print('\n\nInverse of A as ----\n', ai)
```

Eigan Value and Eigan vector 

```
# Eigen Values and Eigen Vectors of a Matrix 
import numpy as np 
import scipy.linalg as la 
n = 4 
P = np.random.randint(0,10,(n,n)) 
print(P) #Create the symmetric matrix S=P.P' 
S = P @ P.T  # @ represents matrix multipllication and .T transpose
print(S) #Let's unpack the eigenvalues and eigenvectors of S 
evals, evecs = la.eig(S) 
print(evals)
 #The eigenvalues all have zero imaginary part and #so they are indeed real numbers:
evals = evals.real 
print(evals) #The corresponding eigenvectors of A are: 
print(evecs) #Let's check that the eigenvectors are orthogonal to each other:
v1 = evecs[:,0] # First column is the first eigenvector print(v1)
v2 = evecs[:,1] # Second column is the second eigenvector print(v2)
print (v1 @v2)
```
